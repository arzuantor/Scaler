#RobustScaler

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df = pd.read_csv('/content/insurance - insurance.csv')
df1=df.copy()
df2=df.copy()
df3=df.copy()
df4=df.copy()
df5=df.copy()
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
rb=RobustScaler()
```

```python
df["age"]=rb.fit_transform(df[["age"]])
```

```python
df.head(5)
```

#using loop in RobustScaler






```python
from pandas.api.types import is_numeric_dtype
```

```python
for col in df1.columns:
  if is_numeric_dtype(df1[col]):
    df1[col]=rb.fit_transform(df1[[col]])
  else:
    df1[col]=le.fit_transform(df1[col])
```

```python
df1.head(2)
```
